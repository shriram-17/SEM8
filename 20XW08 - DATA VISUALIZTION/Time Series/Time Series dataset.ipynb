```python
import pandas as pd
import plotly.express as px
from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv("INFY.csv")
df.head()
```

```python
x = list(df.columns)
x.remove("Close")
x.remove("Adj Close")
x.remove("Date")
x = df[x]
```

```python
x.head()
```

```python
y = df['Close']
y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.30,shuffle=False)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression().fit(X_train,Y_train)
linear_reg.score(X_train,Y_train)
```

```python
Y_pred = list(linear_reg.predict(X_test))
Y_list=list(Y_test)
result = pd.DataFrame({"Close_pred":Y_pred,"Close":Y_list})
```

```python
fig = px.line(x= result['Date'],y = result["Close"],title = 'A simple line graph')
fig.add_trace(px.line(x=result['Date'], y=result["Close_pred"]))
fig.show()
```
